```python
def list_folders(folder_path):
    return [ path.join(folder_path,  folder)  for folder in  next(walk( folder_path), (None, None, []))[1] ]
```

```python
def list_video_files(folder_path):
    return [ path.join(folder_path, file)  for file in next(walk(folder_path), (None, None, []))[2] if path.splitext(file)[1] == '.avi'  ] 
```

```python
def list_image_files(folder_path, extension):
    return [ path.join(folder_path, file)  for file in next(walk(folder_path), (None, None, []))[2] if path.splitext(file)[1] == extension  ] 
```

```python
def get_categories(dataset_location):
    return list(map((lambda x: x.split("\\")[-1]) , list_folders(dataset_location)))
```

```python
def order_video_frames_files(image_paths):
    image_paths.sort(key = lambda x:  int(x.split("\\")[-1].split(".")[0].split("e")[1])  )
    ret = image_paths
    return ret
    
```

```python
def load_image_sequence(folder_path, extension):
    images_path = list_image_files(folder_path, extension)
    images_path = order_video_frames_files(images_path)
    sequence = []
    for image_path in images_path:
        frame = cv2.imread(image_path)
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
        sequence.append(frame)
    video_sequence = np.asarray(sequence)
    return video_sequence
```

```python
def load_videos_category(category_path, category):
    images_folders = list_folders(category_path)
    dataset_chunk_list = []
    for folder_with_video_images in images_folders:
        video_sequence_array = load_image_sequence(folder_with_video_images, ".jpg")
        dataset_chunk_list.append(video_sequence_array)
    dataset_chunk_array = np.asarray(dataset_chunk_list)
    category_array = np.full(shape=len(dataset_chunk_list), fill_value=category)
    return dataset_chunk_array, category_array
        
        
```
